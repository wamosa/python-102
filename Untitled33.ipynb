```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
df = pd.concat([iris.data, pd.Series(iris.target, name='target')], axis=1)
df.head()# this show the first five lines of the dataset
```

```python
df.describe() # it gives the statistical analysis of the  dataset
```

```python
df.info()
```

```python
df.isnull().sum()
```

there are no null values


```python
import matplotlib.pyplot as plt

import seaborn as sns
# Line Chart: Example (not time series, but you can simulate)
df.plot(kind='line', figsize=(8, 5))
plt.title("Mean Features by Species")
plt.ylabel("Value")
plt.show()
```

```python


# Histogram: Distribution of sepal length
sns.histplot(df['sepal length (cm)'], kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.show()
```

```python
sns.barplot(data=df, x='target', y='petal length (cm)', palette='pastel')
plt.xlabel("sepal length")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='target', palette='Set2')
plt.title("Sepal Length vs Petal Length by target")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title='Target')
plt.show()
```
